```python
import pandas as pd

df=pd.read_csv("all-except-activity.csv")

df.head()
```

```python
df.info()
```

```python
#finding missing values and replacing them

col = df.columns
numeric_col = df._get_numeric_data().columns
cat_col = set(col) - set(numeric_col)
cat_col_list  = list(cat_col)

    
for column in cat_col_list:
    df[column] =  df[column].fillna(df[column].mode()[0])
    
df.info()
```

```python
features_selection = ['UserID','Emotion','Label']
features = pd.DataFrame(df, columns = features_selection)
features.head(20)
```

```python
#looking for duplicate data, invalid data or corrupt data and remove it
features.isnull().sum()
```

```python
#Splitting data into train and test sets
from sklearn.model_selection import train_test_split

y = features['Label']
X = features.drop('Label',1)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
#X['Gender'] = labelencoder.fit_transform(X['Gender'])
X['UserID'] = labelencoder.fit_transform(X['UserID'])
X['Emotion'] = labelencoder.fit_transform(X['Emotion'])
y = labelencoder.fit_transform(y)
```

```python
#splitting data into training and test
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.2, random_state = 100 )

```

```python
# Creating the naive bayes classifier object 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score 

model = GaussianNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print ("Accuracy: ", accuracy_score(y_test,y_pred)*100) 
```
